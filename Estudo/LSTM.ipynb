```python
import yfinance as yf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

    # uma função para criar o dataset com base no número de dias a serem observados previamente
def create_dataset(dataset, timesteps=1):
    data_X, data_Y = [], []
    for i in range(len(dataset) - timesteps - 1):
        a = dataset[i:(i + timesteps), 0]
        data_X.append(a)
        data_Y.append(dataset[i + timesteps, 0])
    return np.array(data_X), np.array(data_Y)

#  obtendo dados do yfinance da magalu
stock_data = yf.download('MGLU3.SA', start='2020-01-01', end='2023-01-01')

    # extraindo os preços de fechamento
closing_prices = stock_data['Close'].values.reshape(-1, 1)

  # normalização dos dados
scaler = MinMaxScaler(feature_range=(0, 1))
closing_prices_scaled = scaler.fit_transform(closing_prices)


timesteps = 5 # especificamos para que ele olhe os 5 dias anteriores

              # criando as bases de teste e treinamento
X, y = create_dataset(closing_prices_scaled, timesteps)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Redimensionar os dados para o formato esperado pelo LSTM (amostras, passos de tempo, características)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

            # instanciando e construindo o modelo
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(units=1))

# compilador o modelo usando o optimizador adam
model.compile(optimizer='adam', loss='mean_squared_error')

# treinando o modelo
model.fit(X_train, y_train, epochs=100, batch_size=32)
       # avaliando o modelo
train_loss = model.evaluate(X_train, y_train, verbose=0)
test_loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Treinamento - Loss: {train_loss}')
print(f'Teste - Loss: {test_loss}')

   # gerando as previsões
predicted_prices_scaled = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted_prices_scaled)

 
print(predicted_prices)
```

```python
from sklearn.metrics import mean_squared_error

# Calcular o MSE(mean square error) para os conjuntos de treinamento e teste
train_mse = mean_squared_error(y_train, model.predict(X_train))
test_mse = mean_squared_error(y_test, predicted_prices_scaled)

# Exibir os valores do MSE
print(f'Treinamento - MSE: {train_mse}')
print(f'Teste - MSE: {test_mse}')
```

```python
import plotly.graph_objs as go
y_test_desescalado = scaler.inverse_transform(y_test.reshape(-1, 1))


fig = go.Figure()

fig.add_trace(go.Scatter(x=np.arange(len(y_test_desescalado)), y=y_test_desescalado.flatten(), mode='lines', name='Real'))
fig.add_trace(go.Scatter(x=np.arange(len(predicted_prices)), y=predicted_prices.flatten(), mode='lines', name='Previsto'))

fig.update_layout(title='Preços de fechamento - Real vs Previsto',
                  xaxis_title='Tempo',
                  yaxis_title='Preço de fechamento')

fig.show()
```
